## 1. line plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 50, 40, 45],
}

df = pd.DataFrame(data)

sns.lineplot(x = df.index, y= 'Age',data=df)

plt.title('Age of Individuals')
plt.xlabel('Index')
plt.ylabel('Age')
plt.show()
```

## kernel density estimate 

```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

df['species'] = iris.target
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

sns.kdeplot(data= df[df['species'] == 'virginica']['sepal length (cm)'], label='Virginica', shade=True)

plt.legend()
plt.title('KDE Plot of Sepal Length for Virginica')
plt.xlabel('Sepal Length (cm)') 
plt.ylabel('Density')
plt.show()
```

# customise the seaborn 

## 1. adding titles and axis lables

```python
import seaborn as sns
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt 

iris = sns.load_dataset('iris')

sns.scatterplot(x= 'sepal_length', y= 'sepal_width',data=iris, hue='species')

# add plot tilte and axis labels
plt.title('Scatter Plot ')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Density')
plt.show()
```

## change the styles and grids using seaborm 

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('whitegrid')

sns.boxplot(x='species',y='petal_length', data = sns.load_dataset('iris'))

plt.title('Box Plot of Petal Length by Species')
plt.xlabel('Species')   
plt.ylabel('Petal Length (cm)')
plt.show()
```

## customing color palettes 

## A ) usign built in modules


```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_palette('pastel')

sns.violinplot(x= 'species',y='petal_length', data = sns.load_dataset('iris'))

plt.title('Violin Plot of Petal Length by Species')
plt.xlabel('Species')   
plt.ylabel('Petal Length (cm)')
plt.show()
```

### B ) custom palete 

```python
custom_colors = ['#FF5733', '#33FF57', '#3357FF']

sns.set_palette(custom_colors)

sns.violinplot(x='species', y='petal_length', data=sns.load_dataset('iris'))

plt.title("custom colored petal lenght distribustion")
plt.show()
```
